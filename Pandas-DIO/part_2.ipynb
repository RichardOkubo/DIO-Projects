```python
import os

import pandas as pd
```

```python
df1, df2, df3, df4, df5 = [
    pd.read_excel(os.path.join("./datasets", file))
    for file in (
        "Aracaju.xlsx",
        "Fortaleza.xlsx",
        "Natal.xlsx",
        "Recife.xlsx",
        "Salvador.xlsx")]
```
